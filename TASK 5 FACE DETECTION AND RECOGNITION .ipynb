### An AI application that can detect and recognize faces in images. Use pre-trained face detection models like Haar cascades face detectors.

# import Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

# Load the pre-trained haarcascade face detection model

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
```

# Load the image

```python
image = cv2.imread('test image 2.jpeg')
# convert the image to BGR2RGB scale
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
# convert to gray scale image
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
faces = face_cascade.detectMultiScale(gray)
```

```python
len(faces)
```

```python
# Convert the image to BGR2GRAY scale
img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray)
```

# Detect Faces in the image

```python
faces = face_cascade.detectMultiScale(img_gray)
```

```python
len(faces)
```

# Diplay and Draw around the detected face in the image

```python
# Draw rectangles around the detected faces
for (x, y, w, h) in faces:
    cv2.rectangle(img_gray, (x, y), (x+w, y+h), (255, 0, 0), 2)
# Display the output image with rectangles drawn
plt.imshow(img_gray)
plt.axis('off')  # Hide axes
plt.show()
```
